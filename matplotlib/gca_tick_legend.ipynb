```python
#標題，刻度，添增圖例
#載入相關模組
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 設置 numpy data
x=np.arange(0,12,0.1)
y=np.sin(x)
```

```python
'''
若使用 plt.method，其實就是在呼叫當前的 axes 進行畫圖

matplotlib.pyplot.gcf()
    得到當前的 figure

matplotlib.pyplot.gca()
    得到當前的 axes
    
matplotlib.pyplot.sca(ax)
    設定當前的 axes，並且會自動更正 figure
'''
```

```python
#設置標題

ax1=plt.gca() #先取出當前的軸變量，並且對其進行操作
# 字型樣式: fontname，字體大小: fontsize, 斜體: italic
ax1.set_title("Big Title",fontsize="20",\
    style="italic",fontweight="bold")
ax1.set_xlabel("Days",fontsize="18",fontname="Arial")
ax1.set_ylabel("Quantities",fontsize="18")
```

```python
#設置標題

ax1=plt.gca() #先取出當前的軸變量，並且對其進行操作
# 字型樣式: fontname，字體大小: fontsize, 斜體: italic
ax1.set_title("Big Title",fontsize="20",\
    style="italic",fontweight="bold")
ax1.set_xlabel("Days",fontsize="18",fontname="Arial")
ax1.set_ylabel("Quantities",fontsize="18")

# 設置刻度的標籤
ax1.set_xticks([1,3.5,7,9]) # 設定刻度的位置
ax1.set_xticklabels(["A","L","A","N"]) # 設定刻度的標籤名稱
```

```python
#設置標題

ax1=plt.gca() #先取出當前的軸變量，並且對其進行操作
# 字型樣式: fontname，字體大小: fontsize, 斜體: italic
ax1.set_title("Big Title",fontsize="20",\
    style="italic",fontweight="bold")
ax1.set_xlabel("Days",fontsize="18",fontname="Arial")
ax1.set_ylabel("Quantities",fontsize="18")

# 設置刻度的標籤
ax1.set_xticks([1,3.5,7,9]) # 設定刻度的位置
ax1.set_xticklabels(["A","L","A","N"]) # 設定刻度的標籤名稱

# 刻度的各項操作: 軸變數名稱.tick_params(axis="x/y/both",direction="out/in",color,width,length)
# !!特別注意!! width & length 所設置的參數為數值，不需要放在""內，否則會變成字串
ax1.tick_params(axis="both",direction="in",width=2,\
    color="blue",length=10) 
```

```python
plt.plot(x,y,color="blue",linestyle="-.")
plt.plot(x+2,y+2,color="red",linestyle=":")
```

```python
'''
添加圖例
ax.plot(...,label="名稱")
ax.plot(...,label="名稱")
plt.legend(loc="best/其他位置")
'''
plt.plot(x,y,color="blue",linestyle="-.",label="Wen")
plt.plot(x+0.5,y+0.5,color="red",linestyle=":",label="Tr. me")
plt.legend(loc="lower right") # loc(location):預設值為 best 
```

```python
# 設置線的上下層: zorder=數值，數值越大越上層
plt.plot(x,y,color="blue",linestyle="-.",label="Wen",zorder=1)
plt.plot(x+0.5,y+0.5,color="red",linestyle=":",label="Tr. me",zorder=2)
# 紅線的 zorder=2>藍線 zorder=1，因此紅線在上層
plt.legend(loc="upper right") # loc(location):預設值為 best 
```
