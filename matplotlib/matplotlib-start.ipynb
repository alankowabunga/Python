```python
#載入相關模組
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#numpy data
X=np.arange(0,12,0.1) #每 0.1 為一個單位，設點從 0 ~ 12
Y=np.sin(X)

#畫圖 plt.plot(x象限參數,y象限參數,線的顏色和樣式)
plt.plot(X,Y,"r--")
```

```python
# 冒號 : 線為點點的樣式
plt.plot(X,Y,"r:")
```

```python
# 符號 -. 線為橫配點的樣式
plt.plot(X,Y,"r-.")
```

```python
#可利用 color , linestyle, linewidth 等改變線的樣式
plt.plot(X,Y, color="lime",linestyle="-",linewidth=2)
```

```python
# marker則可以將每個點標示出來，常用的有 O , v , ^
plt.plot(X,Y,color="b",linestyle="-.",marker="o")
```

```python
# marker則可以將每個點標示出來，常用的有 O , v , ^

plt.plot(X,Y,color="orange",linestyle=":",marker="v")
```

```python
# marker則可以將每個點標示出來，常用的有 O , v , ^
plt.plot(X,Y,color="brown",linestyle="--",marker="^")
```

```python
#marker還能分中心顏色 markerfacecolor & 外框顏色 markeredgecolor 兩個操作
plt.plot(X,Y,color="purple",linestyle="-",marker="o",markerfacecolor="black",markeredgecolor="red")
```

```python
# marker還可以調整大小和外框粗細:markersize , markeredgewidth
plt.plot(X,Y,color="black",linestyle="-",markersize=3,markeredgewidth=4,marker="v",markeredgecolor="blue",markerfacecolor="black")
```

```python
#散點圖:兩種方法，1.是在linestyle=""設置為空。2.是使用 plt.scatter(X.Y)
#需要注意的是，若設置 linestyle 為空，必須設置 marker 樣式才會顯示圖表
plt.plot(X,Y,color="black",linestyle="",marker="o")
```

```python
#散點圖:兩種方法，1.是在linestyle=""設置為空。2.是使用 plt.scatter(X.Y)
plt.scatter(X,Y)
```
