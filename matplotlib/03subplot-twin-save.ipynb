```python
'''
子圖 subplot 繪製:
fig,ax=plt.subplots(row_num,col_num)
ax[#].plot(...)
'''
#載入相關模組
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,12,0.1)
y=np.sin(x)

# subplot
fig,ax=plt.subplots(2,1) # ax 代表圖的軸，a[0]即第一張圖，a[1]即第二張圖
ax[0].plot(x,y)
```

```python
'''
畫圖有兩種方法
1. plt.plot()
2. ax[索引值].plot(...) 指定坐標軸
'''
fig,ax=plt.subplots(2,1)
ax[1].plot(x,y)
ax[0].plot(x+2,y+2,"r")
```

```python
'''
設定坐標軸範圍(可做兩圖比較)
ax.set_xlim([#,#])
ax.set_ylim([#,#])
'''
fig,ax=plt.subplots(2,1)
ax[0].plot(x,y)
ax[1].plot(x+2,y+2,"r")

ax[1].set_xlim([0,10])
ax[0].set_xlim([0,10])
```

```python
'''
指數坐標軸
ax.set_xscale('log')
ax.set_xscale('log')
'''
x2=np.arange(1,100,1)
y2=np.exp(x2)

plt.plot(x2,y2)
ax=plt.gca()
ax.set_yscale('log') #y軸變成指數座標軸
```

```python
'''
雙坐標軸: ax.twinx/y()
'''
x2=np.arange(1,100,1)
y2=np.exp(x2)

plt.plot(x2,y2)
ax=plt.gca() #先取出當前的軸變量，並且對其進行操作
ax2=ax.set_yscale('log')

ax2=ax.twinx() # 意思: 共用一個座標軸，畫出來是 y 軸
ax2.plot(x,y,'ro')

ax3=ax2.twiny()
ax3.plot(x,y,color="black",linestyle=":")

# 儲存 plt.savefig('檔名')
plt.savefig("plot drill.png")
```
